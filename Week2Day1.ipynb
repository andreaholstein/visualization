```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL ##install as pillow if needed
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
# making a figure, putting an axes in this, and set the size of the figure
fig, ax = plt.subplots(figsize=(5,3))

# graph, plotting both y vals on same axis, and labelling them with a legend
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = "lower right",
          frameon = True, # adds frame around legend
          fontsize = 12, 
          ncol = 2, # styled labels into multiple cols
          shadow = True) # adds drop shadow to frame, makes frame opaque
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1,label = "Person 1")
ax.plot(x,y2,label = "Person 2")
ax.legend(loc = "upper left",
          bbox_to_anchor = (1,0.5)) # used to anchor our legend; upper left designates corner of legend box that will be pinned to our anchor, 1 = 100% of x axis, 0.5 = 50% of y axis
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")
ax.text(10,95,"Adding a title!", # nums for aligning on the x, and on the y (25,95) less offensive
        ha = "center", # "horizonatal aligment", centered on the coord (10,95) (which is not centered...)
        color = "red",
        size = 20) 
```

```python
fig, ax = plt.subplots()
ax.axis([0,10,0,10])

ax.text(1,5, ". Data: (1,5)", transform = ax.transData) # giving our coordinates based on the values shown on the axes => 1 across, 5 up (straight-forward)
ax.text(0.5,0.1, ". Axes: (0.5,0.1)", transform = ax.transAxes) # giving coordinates in terms of Axes object (graph itself) => 50% across Axes obj, 10% of the way up
ax.text(0.2,0.2, ". Figure: (0.2,0.2)", transform = fig.transFigure) # giving coordinates in terms of Figure object (overall container) => 20% of the way into our Fig obj, 20% of the way up

```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110,180,240,99,220])
```

```python
# 
fig, (ax1, ax2) = plt.subplots(
    ncols = 2,
    nrows = 1,
    figsize = (7,3)
)

font1 = {"family": "serif", "size":10}
font2 = {"family": "serif", "size":14}

#  creating a figure with 2 graphs (i.e. subplots), one with x1,y1 and another with x2,y2
ax1.scatter(x1,y1,
            marker = "o",
            color = "red")
ax2.bar(x2,y2, color = "darkgreen")

ax1.set_title("Scatterplot", fontdict=font2)
ax1.set_ylabel("Measure", fontdict=font1)
ax1.set_xlabel("Time", fontdict=font1)
ax2.set_title("Pirate Strength", fontdict=font2)
ax2.set_ylabel("Strength", fontdict=font1)
ax2.set_xlabel("Names", fontdict=font1)
```

```python
# new fig, with someaxes (subplots)
fig, someaxes = plt.subplot_mosaic([["ax1", "ax3"], # order in here determines subplot layouts
                                    ["ax2", "ax3"]], # so ax3 takes up full height on right side, ax and ax2 stacked on each other on left
                                    figsize=(7,4),
                                    layout = "constrained") # keep labels within fig confines

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)


someaxes["ax1"].set_xlabel("A Big Label", fontsize = 18)
someaxes["ax2"].set_xlabel("Another Label", fontsize = 18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furious", fontsize = 18)
```
